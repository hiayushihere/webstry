```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
main_list = []
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'}

```

```python
def extract(url):
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    articles = soup.find_all('div', class_ = "col-md-10 col-sm-9 col-xs-12 right-section")
    return articles
```

```python
def transform(articles):
    for item in articles:
        name = item.find('span', class_ = 'cmp_name')
        location = item.find('span', {'itemprop': 'streetAddress'})
        try:
            city = item.find('span', class_ = 'locationCity').text
        except:
            city = ''
        try:
            P_O_Box = item.find('span', {'itemprop': 'postalCode'}).text
        except:
            P_O_Box = ''
        try:
            Phone = item.find('span', class_ = 'phonespn').text
        except:
            Phone = ''
        try:
            Mobile = item.find('span', class_ = 'phonespn').text.strip()
        except:
            Mobile = ''
        try:
            company_page_link = item.find('a', class_ = 'website')['href']
        except:
            company_page_link = ''
        try:
            logo_url = item.find('img', {'itemprop': 'image'})['src']
        except:
            logo_url = ''
        business = {
            'name': name,
            'location': location,
            'city': city,
            'P.O Box': P_O_Box,
            'Phone': Phone,
            'Mobile': Mobile,
            'company_page_link':company_page_link,
            'Logo url': logo_url
        }
        main_list.append(business)
    return main_list
```

```python
def load():
    df = pd.DataFrame(main_list)
    df.to_csv('resultt.csv', index=False)
```

```python
for x in range(1,1318):
    print(f'Getting page {x}')
    articles = extract(f'https://www.yellowpages-uae.com/uae/restaurant?page={x}')
    transform(articles)
    time.sleep(5)
```

```python
load()
print('Saved to CSV')
```
